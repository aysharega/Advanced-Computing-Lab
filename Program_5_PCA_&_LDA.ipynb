# Dimensionality Reduction: PCA on load_iris dataset 


```python
# 1. Dataset
from sklearn import datasets
iris = datasets.load_iris()
x=iris.data
y=iris.target
target_names = iris.target_names

# 2. PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pcanew =  pca.fit(x).transform(x)
print(pcanew)

# 3. Graph plotting
import matplotlib.pyplot as plt
xaxis = pcanew[:,0]
plt.xlabel("PC1")
yaxis = pcanew[:,1]
plt.ylabel("PC2")
plt.title("Dimensionality Reduction")
plt.scatter(x=xaxis,y=yaxis,cmap="rainbow")
```

 # Dimensionality Reduction: LDA on load_iris dataset 

```python
# 1. Dataset
from sklearn import datasets
iris = datasets.load_iris()
x=iris.data
y=iris.target
target_names = iris.target_names

# 2. PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
ldanew=lda.fit(x,y).transform(x)
print(ldanew)

# 3. Graph plotting
import matplotlib.pyplot as plt
xaxis = ldanew[:,0]
plt.xlabel("PC1")
yaxis = ldanew[:,1]
plt.ylabel("PC2")
plt.title("Dimensionality Reduction")
plt.scatter(x=xaxis,y=yaxis,cmap="rainbow")
```
