**Hierarchical clustering**


Using different linkages such as single, complete and average linkages.

```python
import numpy as np
from scipy.cluster import hierarchy
ytdist= np.array([ 662. , 877. , 255. , 412. , 996. , 295. ,
468. , 268. , 400. , 754. , 564. , 138. , 219. , 869. , 669. ])
Z= hierarchy.linkage(ytdist, "single")
output= hierarchy.dendrogram(Z)
```

Using load_iris dataset

```python
#dataset
from sklearn.datasets import load_iris
bc = load_iris()
x = bc.data
Z= hierarchy.linkage(x, "weighted")
output= hierarchy.dendrogram(Z)
```

Using Wine dataset

```python
# wine dataset
from sklearn.datasets import load_wine
wine = load_wine()
x= wine.data
Z= hierarchy.linkage(x, "centroid", optimal_ordering=True)
output= hierarchy.dendrogram(Z)
```
